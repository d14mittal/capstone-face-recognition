# Data Collection

```python
import cv2

classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def extractor(image):
    
    
    faces = classifier.detectMultiScale(image, 1.3, 5)
    
    if faces is ():
        return None
    
    for (x,y,w,h) in faces:
        x=x-10
        y=y-10
        new_cropped_face = image[y:y+h+50, x:x+w+50]

    return new_cropped_face

capture = cv2.VideoCapture(0)
counter = 0

while True:

    return_value, frame_captured = capture.read()
    if extractor(frame_captured) is not None:
        counter += 1
        face = cv2.resize(extractor(frame_captured), (400, 400))

        file_path = './Dataset/Test/New Folder/' + str(counter) + '.jpg'
        cv2.imwrite(file_path, face)

        cv2.putText(face, str(counter), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        
    else:
        print("Face was not found")
        pass

    if cv2.waitKey(1) == 13 or counter == 100:
        break
        
capture.release()
cv2.destroyAllWindows()      
print("Collecting Images Complete")
```
